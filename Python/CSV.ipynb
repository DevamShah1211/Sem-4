```python
import pandas as pd
import matplotlib.pyplot as plt
x=pd.read_csv("auto-mpg.csv")
y=pd.DataFrame(x)
pd.plotting.boxplot(x["mpg"])
q1=y["mpg"].quantile(0.25)
q3=y["mpg"].quantile(0.75)
IQR=q3-q1
low=(q1-1.5*IQR)
high=(q3+1.5*IQR)
out=y[(y["mpg"]<low)|(y["mpg"]>high)]
out
```
