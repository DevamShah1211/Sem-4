```python
import matplotlib.pyplot as plt
import pandas as pd
x=pd.read_csv("auto-mpg.csv")
y=pd.DataFrame(x)
y
```

```python
plt.fill_between(y["model year"],y["cylinders"],y["mpg"],color="red")
plt.plot(y["model year"],y["cylinders"],y["mpg"],label=["Year","Cylinders","MPG"])
plt.show()
```

```python
plt.stackplot(y["model year"],y["cylinders"],y["mpg"],labels=["Cylinders","MPG"])
plt.legend()
plt.show()
```

```python
plt.fill_between(y["mpg"],y["cylinders"]>4,color="red")
plt.fill_between(y["mpg"],y["cylinders"]<4,color="blue")
plt.plot(y["cylinders"],label=["Cylinders"])
plt.show()
```

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
g=nx.Graph()
g.add_nodes_from(["A","B","C","D","E"])
g.add_edges_from([("A","B"),("A","C"),("A","D"),("A","E"),("B","C"),("B","D"),("B","E"),("C","B"),("C","D"),("C","E"),("D","B"),("D","C"),("D","E"),("E","B"),("E","C"),("E","D")])
nx.draw(g,with_labels=True)
plt.show()
```

```python
ins={"P1":[1,1,0,1,1],"P2":[1,1,1,0,1],"P3":[1,1,1,1,0],"P4":[0,1,1,1,1],"P5":[1,0,1,1,1]}
g=nx.DiGraph()
g.add_nodes_from(ins.keys())
for i,j in enumerate(ins):
    for k,h in enumerate(ins[j]):
        if h:
            g.add_edge(j,list(ins.keys())[k])
nx.draw(g,with_labels=True,node_size=1000,edge_color="red")
plt.show()
```

```python
from wordcloud import WordCloud,STOPWORDS
f=open("Python.txt","r")
x=f.read()
st=set(STOPWORDS)
y=WordCloud(background_color="black",max_words=200000,stopwords=st)
y.generate(x)
plt.imshow(y)
plt.axis("off")
plt.show()
```

```python
from wordcloud import WordCloud,STOPWORDS
import requests as rq
# f=open("Python.txt","r")
# x=f.read()
f="https://en.wikipedia.org/wiki/Python_(programming_language)"
x=rq.get(f)
tx=x.text
st=set(STOPWORDS)
y=WordCloud(background_color="black",max_words=200000,stopwords=st)
y.generate(tx)
plt.imshow(y)
plt.axis("off")
plt.show()
```
