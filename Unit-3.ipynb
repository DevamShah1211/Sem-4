```python
import seaborn as sns
import networkx as nx
import matplotlib.pyplot as plt
g=nx.Graph()
node=[1,2,3,4,5]
edge=[(1,1),(1,2),(1,3),(1,4),(1,5),(2,2),(2,3),(2,4),(2,5)]
g.add_nodes_from(node)
g.add_edges_from(edge)
nx.draw(g,with_labels=True)
plt.show()
```

```python
import seaborn as sns
import networkx as nx
import matplotlib.pyplot as plt
ins={"p1":[1,1,1,0,1],"p2":[0,0,1,0,1],"p3":[1,1,0,1,1],"p4":[1,1,1,0,0],"p5":[1,1,0,0,1]}

g=nx.DiGraph()
g.add_nodes_from(ins.keys())

for i,j in enumerate(ins):
    for k,h in enumerate(ins[j]):
        if h==1:
            g.add_edge(j,list(ins.keys())[k])
            
nx.draw(g,with_labels=True,node_size=850,edge_color="red")
plt.show()
```
