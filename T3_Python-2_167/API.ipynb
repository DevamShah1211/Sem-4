```python
import json,requests
lat=23.0225
lon=72.5714
API_key="0b9ecbf3f0be88096f138a380ba86af8"
data=f"https://api.openweathermap.org/data/2.5/forecast?lat={lat}&lon={lon}&appid={API_key}"
x=requests.get(data)
z=x.json()
print(json.dumps(z,indent=5))
```

```python
dt=[]
weather=[]
temp=[]
press=[]
humid=[]
speed=[]
for i in z["list"]:
    dt.append(i["dt_txt"])
    temp.append(i["main"]["temp"])
    press.append(i["main"]["pressure"])
    humid.append(i["main"]["humidity"])
    weather.append(i["weather"][0]["description"])
    speed.append(i["wind"]["speed"])
```

```python
import pandas as pd
data={"Date-Time":dt,"Temperature":temp,"Pressure":press,"Humidity":humid,"Weather":weather,"Speed":speed}
df=pd.DataFrame(data)
import numpy as np
df["Alert"]=np.where(df["Speed"]>3,"Red Alert","Green Alert")
df
```
